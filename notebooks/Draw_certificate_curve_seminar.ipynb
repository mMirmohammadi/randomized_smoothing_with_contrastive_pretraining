```python
# for name in dfs.keys():
#     d = dfs[name]
#     ACR = d[(d['correct'] == 1)]['radius'].sum() / len(d)
#     print(f"{name} | ACR: {100 * ACR:.2f}")

# Make a dataframe with each row for different run and columns as follow: [name, ACR, radius 0.0, radius 0.25, radius 0.5, radius 0.75, radius 1.0, radius 1.25, radius 1.5]

results = pd.DataFrame(columns=['name', 'num', 'ACR', 'STD_CR', 'radius 0.0', 'radius 0.25', 'radius 0.5', 'radius 0.75', 'radius 1.0', 'radius 1.25', 'radius 1.5', 'radius 1.75', 'radius 2.0', 'radius 2.25'])

def get_acc_at_radius(d, radius):
    return (d["correct"] & (d["radius"] >= radius)).mean() * 100

for name in dfs.keys():
    d = dfs[name]
    ACR = d[(d['correct'] == 1)]['radius'].sum() / len(d)
    STD_CR = d[(d['correct'] == 1)]['radius'].std()
    tmp = pd.DataFrame({
        'name': name,
        # 'name': name.split(' |')[0],
        # 'arc': name.split('| ')[1].split(' ')[0],
        # 'sigma': float(name.split('$')[-1]),
        'num': len(d),
        'STD_CR': STD_CR,
        'ACR': ACR,
        'radius 0.0': get_acc_at_radius(d, 0.0),
        'radius 0.25': get_acc_at_radius(d, 0.25),
        'radius 0.5': get_acc_at_radius(d, 0.5),
        'radius 0.75': get_acc_at_radius(d, 0.75),
        'radius 1.0': get_acc_at_radius(d, 1.0),
        'radius 1.25': get_acc_at_radius(d, 1.25),
        'radius 1.5': get_acc_at_radius(d, 1.5),
        'radius 1.75': get_acc_at_radius(d, 1.75),
        'radius 2.0': get_acc_at_radius(d, 2.0),
        'radius 2.25': get_acc_at_radius(d, 2.25),

    }, index=[0])
    results = pd.concat([results, tmp], ignore_index=True)

# sort by sigma, then by name
results = results.sort_values(by=['ACR'])

# results.to_csv('results.csv', index=False)

results.round(3)
    
```
